```python
import numpy as np, statistics
from matplotlib import pyplot as plt
```

```python
data = [[1999,5427, 18.079],
[2000,5688, 18.594],
[2001,6198, 19.753],
[2002,6462, 20.734],
[2003,6635, 20.831 ],
[2004,7336, 23.029],
[2005,7248, 23.597],
[2006,7491, 23.584],
[2007,8161, 25.525 ],
[2008,8578, 27.731],
[2009,9000, 29.449]]
data = np.array(data)
year = data[:,0]
suicides = data[:,1]/1000
scale = 8/25
science = data[:,2] * scale
def create_plot(ticks=None):
    fig = plt.figure(figsize=(15,5))
    ax = fig.add_subplot(1,1,1)
    if ticks is not None: ax.set_xticks(ticks)
    return ax
```

```python
ax = create_plot(year)
ax.plot(year,suicides,'r-')
ax.plot(year,science, 'g-')
```

```python
ax = create_plot()
ax.plot(suicides,science,'o')
```

```python
statistics.correlation(suicides,science)
```

```python
ax = create_plot(year)
ax.plot(year,15-suicides, 'r-')
ax.plot(year,science, 'g-')
```

```python
ax = create_plot()
ax.plot(15-suicides,science,'o')
```

```python
statistics.correlation(15-suicides,science)
```

```python
ax = create_plot(year)
np.random.shuffle(science)
ax.plot(year,suicides, 'r-')
ax.plot(year,science, 'g-')
```
